```python
data = {"Name": ["Alice", "Bob"], "Age": [25, 30]}
df = pd.DataFrame(data)
print(df)
```

```python
print(df.info())
print("----------")
print(df.describe())
```

```python

print(df[["Name", "Age"]])
```

```python

print(df[df["Age"] > 25])
```

```python
print(df.loc[0])
print(df.loc[:, "Name"])
```
